```python
# Independent and Dependent Features 
X=df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
Y=df['total_bill']
```

```python
# Train and Test Data
from sklearn.model_selection import train_test_split
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,test_size=0.25,random_state=10)
```

```python
Xtrain
```

```python
## Features Encoding: Label Encoding and OneHot Encoding
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
```

```python
import warnings
warnings.filterwarnings('ignore')
Xtrain['sex']=le1.fit_transform(Xtrain['sex'])
Xtrain['smoker']=le2.fit_transform(Xtrain['smoker'])
Xtrain['time']=le3.fit_transform(Xtrain['time'])
```

```python
Xtrain.head()
```

```python
Xtest['sex']=le1.transform(Xtest['sex'])
Xtest['smoker']=le2.transform(Xtest['smoker'])
Xtest['time']=le3.transform(Xtest['time'])
```

```python
Xtest.head()
```

```python
# One Hot Encoding--------ColumnTransformer

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct=ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop='first'),[3])],
                    remainder='passthrough')
```

```python
import sys
import numpy as np
np.set_printoptions(threshold=sys.maxsize)
Xtrain=ct.fit_transform(Xtrain)
```

```python
Xtest=ct.fit_transform(Xtest)
```

## SVR - SUPPORT VECTOR REGRESSION

```python
from sklearn.svm import SVR
```

```python
svr=SVR()
```

```python
svr.fit(Xtrain,Ytrain)
```

```python
Ypred=svr.predict(Xtest)
```

```python
# R^2 and Mean Absolute Error Metrics
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
r2=r2_score(Ytest,Ypred)
mae=mean_absolute_error(Ytest,Ypred)
print(f"R^2 Score: {r2}")
print(f"MAE: {mae}")
```

```python
## Hyperparameter tuning 
from sklearn.model_selection import GridSearchCV

#defining parameter range 
paramGrid={'C': [0.1,1,10,100,1000],
          'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
          'kernel': ['linear','rbf']}
```

```python
grid=GridSearchCV(SVR(),param_grid=paramGrid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(Xtrain,Ytrain)
```

```python
grid.best_params_
```

```python
## Predicition
gridPred=grid.predict(Xtest)
```

```python
r2=r2_score(Ytest,gridPred)
mae=mean_absolute_error(Ytest,gridPred)
print(f"R^2 Score: {r2}")
print(f"MAE: {mae}")
```
