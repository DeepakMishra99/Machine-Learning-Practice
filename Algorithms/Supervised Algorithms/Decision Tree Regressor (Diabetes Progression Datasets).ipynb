```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
```

```python
dataset=load_diabetes()
```

```python
dataset
```

```python
print(dataset.DESCR)
dataset.feature_names
```

```python
# Create DataFrame
df_diabetes=pd.DataFrame(dataset.data,columns=['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6'])
```

```python
X=df_diabetes
Y=dataset['target']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.30,random_state=10)
```

```python
X_train
Y_train
```

```python
# Visulaization
X_train.corr()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(X_train,Y_train)
```

# Hyperparameter Tuning

```python
params={
    'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5,10,15,20,25,30],
    'max_features':['sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(regressor,param_grid=params,cv=5,scoring='neg_mean_squared_error')
```

```python
grid
```

```python
grid.best_params_
```

```python
Y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
r2=r2_score(Y_test,Y_pred)
mse=mean_squared_error(Y_test,Y_pred)
mae=mean_absolute_error(Y_test,Y_pred)
print("R^2 Score",r2)
print("Mean Squared Error",mse)
print("Mean Absolute Error",mae)
```

```python
## Model Selection

modelSelected=DecisionTreeRegressor(criterion='squared_error',max_depth=3,max_features='log2',splitter='best')
```

```python
modelSelected.fit(X_train,Y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(modelSelected,filled=True)
```
