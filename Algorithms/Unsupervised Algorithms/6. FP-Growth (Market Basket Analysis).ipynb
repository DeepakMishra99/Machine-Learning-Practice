### Importing the basic libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing the dataset

```python
dataset = pd.read_csv('Datasets/market_basket_data.csv',header=None)
dataset.head()
```

```python
dataset.shape

```

### Data Pre-Processing

```python
transactions = []
for i in range(0, 7500):
  transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
transactions
```

### Training FP-Growth

```python
!pip install mlxtend
```

```python

from mlxtend.preprocessing import TransactionEncoder
encoder=TransactionEncoder()


l1=encoder.fit(transactions).transform(transactions)
dt=pd.DataFrame(l1,columns=encoder.columns_)
dt.drop(columns=['nan'], inplace=True)
dt
```

```python
from mlxtend.frequent_patterns import fpgrowth
dt.fillna(0, inplace=True)
res=fpgrowth(dt,min_support=0.0045,use_colnames=True,max_len=2)
res.dropna(inplace=True)
res.head(30)
```

```python
from mlxtend.frequent_patterns import association_rules
res=association_rules(res,metric='confidence',min_threshold=0.2,num_itemsets=float)
res
```

```python
res.sort_values("confidence",ascending=False)
```
