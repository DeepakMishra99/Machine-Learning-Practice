```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris_dataset=load_iris()
```

```python
iris_df=pd.DataFrame(iris_dataset.data,columns=iris_dataset.feature_names)
iris_df
```

```python
print(iris_dataset.target_names)
```

```python
iris_df.isna().sum(),iris_df.isnull().sum()
```

```python
# Standard Scaler
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X=scaler.fit_transform(iris_df)
```

```python
Y=(iris_dataset['target'])
```

```python
Y
```

```python
#Apply PCA
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
```

```python
X_pca=pca.fit_transform(X)
```

```python
X_pca
```

```python
plt.scatter(X_pca[:,0],X_pca[:,1],c=Y)
```

```python
# Apply Agglomerative Clustering
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,7))
plt.title("Iris Agglomerative Clustering")

sc.dendrogram(sc.linkage(X_pca,method='ward'))
plt.xlabel("Sample Index")
plt.ylabel("Eucledian Distance")
```

## When using Agglomerative Clustering with PCA on the Iris dataset, it's not unusual to see only 2 clusters emerge, even though the dataset has 3 species. This happens because PCA reduces the dimensionality, often capturing only the most dominant patterns in the data, which may group certain classes together in lower-dimensional space.

```python
## Agglomerative Clustering 

from sklearn.cluster import AgglomerativeClustering
```

```python
cluster=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
```

```python
cluster.fit(X_pca)
```

```python
cluster.labels_
```

```python
plt.scatter(X_pca[:,0],X_pca[:,1],c=cluster.labels_)
```
